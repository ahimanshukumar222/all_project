```python
# First I will extract all my followers and will add to my_followers and then extract all the mutual  and will add to
# mutual_followers and then will do the difference set so as to get the followers who are following foodtalkindia but not me

my_followers = set()
mutual_followers = set()
search = driver.find_element(By.XPATH,'//*[@id="react-root"]/section/nav/div[2]/div/div/div[2]/input')
search.send_keys('username')
sleep(3)
page = driver.find_element_by_class_name('-qQT3')
page.click()
sleep(3)
follower_btn = driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/header/section/ul/li[2]/a/div')
follower_btn.click()

pop_up_window = WebDriverWait(driver, 2).until(EC.element_to_be_clickable((By.XPATH, "//div[@class='isgrP']")))
i=1
while True:
    i+=1
    if i==90:
        break
    driver.execute_script(
        'arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].offsetHeight;', 
      pop_up_window)
    time.sleep(1)
a = driver.find_elements_by_xpath('//span[starts-with(@class,"Jv7Aj")]')
for e in range(290):
    username = a[e].get_attribute('outerHTML').split(' ')[9].split('=')[1].strip()
    my_followers.add(username.strip('""'))
    
sleep(2)
driver.back()
sleep(2)
driver.back()
sleep(2)

search = driver.find_element(By.XPATH,'//*[@id="react-root"]/section/nav/div[2]/div/div/div[2]/input')
search.send_keys('foodtalkindia')
sleep(3)
page = driver.find_element_by_class_name('-qQT3')
page.click()
sleep(5)
accounts=driver.find_element_by_xpath('//div[contains(@class,"PIoXz")]')
accounts.click()
sleep(3)
see_all = driver.find_element_by_xpath('/html/body/div[6]/div/div/div/div[2]/div/a')
see_all.click()
a = driver.find_element_by_xpath('//span[starts-with(@class,"Jv7Aj")]')
print(a.get_attribute('outerHTML'))
a = driver.find_elements_by_xpath('//span[starts-with(@class,"Jv7Aj")]')
for e in range(3):
    username = a[e].get_attribute('outerHTML').split(' ')[9].split('=')[1].strip()
    mutual_followers.add(username.strip('""'))
sleep(2)
for ele in mutual_followers.difference(my_followers):
    print(ele)

```
